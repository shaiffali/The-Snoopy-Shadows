```python
"""
Submitted on Tue Oct 19 2020 23:40 
@author: Shaiffali Sagar
"""

""" 
Introduction:
    'The Snoopy Shadows' - the investigator agency is looking for really smart, intelligent and sharp-witted detectives.
    The game challenges the inner sleuths of the players to solve 5 set of mysteries.
    The player who solves all the 5 mysteries will be hired for assisting,the Great Sherlock Holmes.


    Rules:
    1. The Player will get 2 helplines that can be used anytime throughout the game.
    2. The player may choose to quit the game anytime.
    3. While selecting a Helpline, Use has to input 'Helpline' first and then input the helpline option selected.

    NOTE: Please keep the file: question_bank.txt in the same directory


    Known Bugs and/or Errors:
    None
"""

# Importing python's module
import os
import time
import random 
from colorama import Fore, Style
import matplotlib.pyplot as plt

# reading the questions from a file
FILENAME="question_bank.txt"

#function for helpline: removes 2 incorrect options
def use_5050(options, answer):
    return random.choice([i for i in range(4) if options[i] != answer])

# function for helpline: displays the bar of correct answers from the Experts              
def use_poll(quest,options,answer,helpline):
    #Generate random numbers that sum to 100
    nums = random.sample(range(1,100), len(options))
    def divide(num):
        return round(num/sum(nums)*100,2) 
        
    #create a list of the random numbers generated    
    Percent = list(map(divide, nums))
    
    #display plot of Bars
    #ax.bar(x, height, width, bottom, align)
    chart = plt.figure()
    quadrant = chart.add_axes([0,0,1,1])
    quadrant.bar(options,Percent,color = 'g')
    print('The Experts Poll says:')
    plt.show()

#process the user input if helpline is selected
def process_helpline(quest,options,answer,helpline):
    if not helpline:
        print("You have used all of your helplines!!\n")
        return None
    print(f"""\n\t\t{Fore.GREEN}{Style.BRIGHT}HelpLines: {list(helpline.keys())}\t{Style.RESET_ALL}\n""")
    helpline_chosen = input(prompt = "Which HelpLine would you like to use?? ")
    helpline_chosen = helpline_chosen.upper()
    if helpline_chosen in helpline:
        return helpline_chosen
    return None
    
#process the user input if answer is selected
def process_answer(user_input, answer):     
    if user_input.upper() == answer.upper():
        return True
    return False
        
#prints the question with options
def display_question(quest,options,answer,helpline):
    time.sleep(1)
    print(f"""\n{Fore.RED}{Style.BRIGHT}{quest}{Style.RESET_ALL}\n""")
    for opt in options:
        print(f"""{Fore.RED}{Style.BRIGHT}\t* {opt}{Style.RESET_ALL}""")
    time.sleep(1)
    
#collects user inputs    
def get_user_input(quest,options,answer,helpline):
    user_input = input(prompt = 'Please type in the COMPLETE OPTION for the answer or choose HELPLINE or choose QUIT: ')
    user_input = user_input.upper()
    print(f'''\n\t Alright.. You have decided to go for {user_input}!''')
    time.sleep(2)
    return user_input

#reads the data from the file
def get_question_from_file():
    fp = open(FILENAME, "r")
    questions = []
    for line in fp:
        data = line.split("#")
        answer = data[5].strip("\n")
        questions.append([data[0], [data[1], data[2], data[3], data[4]], answer])
    return questions

#sorts the data read from file into questions,options & answers 
def get_question_details(mystery, round_number):
    r1 = random.randint(0, len(mystery)-1)
    #display quest1
    rnd1 = mystery[r1]
    quest = rnd1[0]
    options = rnd1[1]
    answer = rnd1[2]
    
    mystery.pop(r1)
    return quest,options,answer  
    
#Main function call    
def start_game():
    print(f"""Hello!!\U0001F642 I am Shaiffali Sagar, the Hiring Manager for 'The Snoopy Shadows' \U0001F9D0
As an investigator agency, 'The Snoopy Shadows' is looking for really smart and sharp-witted detectives,
who will be assisting The Great Sherlock Holmes for his next cases...\n Exciting.. isn't it??\n\n""")
    time.sleep(1)
    print(f"""To join the agency, All you need is Intelligence,Analysis skills and a Clever Name!!
As a hiring process:
\t 1. The candidate will be tested for solving a set of 5 murder mysteries.
\t 2. The candidate will get 2 helplines that can be used anytime throughout the game:
\t\t  50-50: removes 2 incorrect options
\t\t   POLL: displays the percentage of correct answers from the Experts
\t 3. While selecting a Helpline, candidate has to input 'Helpline' first and then input the helpline option selected.
\t 4. The player may choose to quit the game anytime.\n""")
    time.sleep(2)
    
    print('So, Here is your chance to challenge your inner Sleuth and apply your quick-wit to solve the mysteries!!')
    
    candidate_name = input(prompt="What's your Clever name: ")
    print(f"\nWelcome {candidate_name.capitalize()}!!!!\n")
    time.sleep(1)
    input('<Press ENTER to start with your 1st challenge> ')
    
    #helpline added as dictionary
    helpline = {'50-50': True, 'POLL': True}
    #list of questions read from file
    mystery = get_question_from_file() 
    count = 1
    next_round = False
    won = True
    quest,options,answer = get_question_details(mystery,count)
    #running the loop for 5 rounds of mysteries
    while count <= 5:
        print("*" * 42)
        print(f"\t\t Round: {count}")
        print("*" * 42)
        if next_round:
            quest,options,answer = get_question_details(mystery,count)
            next_round = False
        display_question(quest, options, answer, helpline)
        user_input = get_user_input(quest, options, answer, helpline)
        if user_input.upper() == 'HELPLINE':
            helpline_chosen = process_helpline(quest, options, answer, helpline)
            if helpline_chosen is not None:
                if helpline_chosen == '50-50':
                    helpline.pop('50-50')
                    opt1 = use_5050(options, answer)
                    options = [options[opt1], answer]
                elif helpline_chosen.upper() == 'POLL':
                    helpline.pop('POLL')
                    use_poll(quest, options, answer, helpline)
                else:
                    print("Choose correct Helpline or select your answer...")
                continue
            else:
                print("Helpline doesn't exists!! Please change your input")
        elif user_input.upper() == 'QUIT':
            print('\n\tThanks for applying!!\U0001F642 GoodLuck!!\n')
            won = False
            break
        else:
            if process_answer(user_input,answer) is True:
                count +=1
                print(f"""\n\t'{user_input}' is absolutely correct!!""")
                next_round = True
            else:
                print('\n\tOOPS!! Wrong Selection!!\U0001F610\n\tBetter Luck next time!!\U0001F642')
                won = False
                break
    if won:
        print("\n**** \U0001F60ACongratulations\U0001F60A ****\nYou've made it to 'The Snoopy Shadows'!!\U0001F9D0")
```

```python
start_game()
```
